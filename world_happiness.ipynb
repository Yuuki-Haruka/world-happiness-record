## Part A: Data Analysis on Year 2019



This project analyzes the *World Happiness Report 2019* dataset obtained from Kaggle.

### 1. Upload Data




The dataset used in this project is the *World Happiness Report 2019*, obtained from Kaggle. 
It contains happiness scores and contributing factors for different countries around the world. 
The dataset was loaded using Pandas, and the first few rows were displayed to understand 
its structure and key variables.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('World_Happiness_Records/2019.csv')
df.head()
```

### 3. Data Cleaning


Data cleaning was performed to improve the accuracy and reliability of the analysis.
This included handling missing values and removing duplicate country records.
Cleaned data ensures valid comparisons between countries and happiness-related factors
in the World Happiness Report 2019.


#### 3.1 Handle Missing Values



Missing values were examined in the World Happiness Report 2019 dataset to assess data quality.
Identifying missing values is important to ensure that happiness scores and contributing factors
are analyzed accurately and do not lead to misleading conclusions.


```python
df.isnull().sum()
```

```python
numeric_cols=df.select_dtypes(include='number').columns
df_cleaned=df.copy()
df_cleaned[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

#### 3.2 Remove Duplicated Data



Duplicate country or region entries were checked to ensure that each country appears only once
in the World Happiness Report 2019 dataset. Removing duplicate records helps maintain the
reliability of happiness rankings and comparisons across countries.


```python
print(df_cleaned.duplicated().sum())
df_cleaned=df_cleaned.drop_duplicates()
```

Histograms were used to visualize the distribution of the happiness score and its contributing factors

These plots help identify how happiness-related variables are distributed across countries in 2019.

## Part B: Multi-Year Trend Analysis (2015-2019)



In addition to the detailed analysis of the 2019 dataset, this section provides a
high-level trend analysis using World Happiness Report data from 2015 to 2019.
The goal is to observe overall changes in global happiness over time rather than
performing a detailed country-level analysis for each year.


### 1. Upload Multiple Files

```python
df_2015=pd.read_csv('World_Happiness_Records/2015.csv')
df_2016=pd.read_csv('World_Happiness_Records/2016.csv')
df_2017=pd.read_csv('World_Happiness_Records/2017.csv')
df_2018=pd.read_csv('World_Happiness_Records/2018.csv')
df_2019=pd.read_csv('World_Happiness_Records/2019.csv')
```

### 2. Standardize and Arrange Column Names

The World Happiness datasets from different years use inconsistent column names for the same concepts. For example, happiness score and GDP per capita appear under different labels such as Happiness Score, Score, or Economy (GDP per Capita).
To enable reliable merging, aggregation, and trend analysis across multiple years, all column names are standardized into a common schema.

```python
df_2015 = df_2015.rename(columns={"Country": "Country",
    "Happiness Score": "score",
    "Economy (GDP per Capita)": "gdp",
    "Family": "social_support",
    "Health (Life Expectancy)": "healthy_life",
    "Freedom": "freedom",
    "Generosity": "generosity",
    "Trust (Government Corruption)": "corruption"})
df_2016 = df_2016.rename(columns={"Country": "Country",
    "Happiness Score": "score",
    "Economy (GDP per Capita)": "gdp",
    "Family": "social_support",
    "Health (Life Expectancy)": "healthy_life",
    "Freedom": "freedom",
    "Generosity": "generosity",
    "Trust (Government Corruption)": "corruption"})
df_2017 = df_2017.rename(columns={"Country": "Country",
    "Happiness.Score": "score",
    "Economy..GDP.per.Capita.": "gdp",
    "Family": "social_support",
    "Health..Life.Expectancy.": "healthy_life",
    "Freedom": "freedom",
    "Generosity": "generosity",
    "Trust..Government.Corruption.": "corruption"})
df_2018 = df_2018.rename(columns={"Country or region": "Country",
    "Score": "score",
    "GDP per capita": "gdp",
    "Social support": "social_support",
    "Healthy life expectancy": "healthy_life",
    "Freedom to make life choices": "freedom",
    "Generosity": "generosity",
    "Perceptions of corruption": "corruption"})
df_2019 = df_2019.rename(columns={"Country or region": "Country",
    "Score": "score",
    "GDP per capita": "gdp",
    "Social support": "social_support",
    "Healthy life expectancy": "healthy_life",
    "Freedom to make life choices": "freedom",
    "Generosity": "generosity",
    "Perceptions of corruption": "corruption"})

df_2015["Year"]=2015
df_2016["Year"]=2016
df_2017["Year"]=2017
df_2018["Year"]=2018
df_2019["Year"]=2019

cols=["Country", "Year", "score", "gdp", "social_support","healthy_life", "freedom", "generosity", "corruption"]

df_2015=df_2015[cols]
df_2016=df_2016[cols]
df_2017=df_2017[cols]
df_2018=df_2018[cols]
df_2019=df_2019[cols]

df_all=pd.concat([df_2015, df_2016, df_2017, df_2018, df_2019],ignore_index=True)
mean_by_year=(df_all.groupby("Year", as_index=False).mean(numeric_only=True))
mean_by_year

```

### 3. Data Visualization

In this section, data visualization techniques are used to explore and compare happiness scores and related factors across different years and countries. Visualizations help transform numerical data into clear and interpretable patterns, making trends and differences easier to understand.


#### 3.1 Line Plot: Happiness Score 

```python
mean_year=df_all.groupby("Year")["score"].mean().reset_index()
mean_year
```

This line plot compares the average, highest, and lowest happiness scores for each year.
The highest happiness scores remain consistently above 7.5, while the lowest scores stay below 3.0, highlighting a significant gap between the happiest and least happy countries. The average score lies steadily between these extremes, emphasizing inequality in happiness levels across countries.

```python
max_year=df_all.groupby("Year")["score"].max().reset_index()
min_year=df_all.groupby("Year")["score"].min().reset_index()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(mean_year["Year"], mean_year["score"], marker="o", linewidth=2, markersize=8, label="Average Score")
plt.plot(max_year["Year"], max_year["score"], marker="^", linestyle="--", linewidth=2, label="Highest Score", color='green')
plt.plot(min_year["Year"], min_year["score"], marker="v", linestyle="--", linewidth=2, label="Lowest Score", color='red')
plt.xlabel("Year", fontsize=11)
plt.ylabel("Happiness Score", fontsize=11)
plt.title("Average vs Highest vs Lowest Happiness Scores Over Time", fontsize=14)
plt.ylim(0, 8)
plt.yticks([3, 4, 5, 6, 7, 8])
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

#### 3.2 Line Plot: Top 5 Countries Comparision

This line plot tracks the happiness scores of the top five happiest countries over time.
The visualization shows that these countries consistently maintain high happiness scores, with Finland showing a noticeable increase and becoming the highest-ranked country in later years.

```python
top5 = (df_all.groupby("Country")["score"].mean().sort_values(ascending=False).head(5).index)
df_top5 = df_all[df_all["Country"].isin(top5)]
plt.figure(figsize=(14,6))
sns.lineplot(data=df_top5, x="Year", y="score", hue="Country", marker="o")
for _, row in df_top5.iterrows():
    plt.text(row["Year"],row["score"],f"{row['score']:.2f}",fontsize=8,ha="center",va="bottom",bbox=dict(facecolor="white", edgecolor="none", alpha=0.7))
plt.xlabel("Year")
plt.ylabel("Happiness Score")
plt.title("Top 5 Happiest Countries Over Time")
plt.show()
```
